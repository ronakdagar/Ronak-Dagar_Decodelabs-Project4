```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Dataset
file_path = "vehicle_dataset.csv"   
df = pd.read_csv("C:/Users/ronak/OneDrive/Desktop/vehicle_dataset.csv")
```

```python
# Display First Rows
print("Dataset Preview:")
display(df.head())
```

```python
# Dataset Information
print("\nDataset Info:")
print(df.info())
```

```python
# Basic Statistics
print("\nSummary Statistics:")
display(df.describe())
```

```python
# 1. BAR CHART - Count of Vehicles by Brand
top_brands = df['Car_Name'].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_brands.index, y=top_brands.values)
plt.title("Top 10 Vehicle Brands/Models")
plt.xlabel("Vehicle")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. HISTOGRAM - Selling Price Distribution
plt.figure(figsize=(10,6))
sns.histplot(df['Selling_Price'], bins=20, kde=True)
plt.title("Distribution of Vehicle Selling Price")
plt.xlabel("Selling Price (Lakhs)")
plt.ylabel("Frequency")
plt.show()
```

```python
# 3. HISTOGRAM - Mileage Distribution
plt.figure(figsize=(10,6))
sns.histplot(df['Kms_Driven'], bins=20, kde=True)
plt.title("Distribution of Vehicle Mileage")
plt.xlabel("Kilometers Driven")
plt.ylabel("Frequency")
plt.show()
```

```python
# 4. BOX PLOT - Price by Fuel Type
plt.figure(figsize=(10,6))
sns.boxplot(x='Fuel_Type', y='Selling_Price', data=df)
plt.title("Selling Price by Fuel Type")
plt.xlabel("Fuel Type")
plt.ylabel("Selling Price")
plt.show()
```

```python
# 5. SCATTER PLOT - Mileage vs Price
plt.figure(figsize=(10,6))
sns.scatterplot(
    x='Kms_Driven',
    y='Selling_Price',
    hue='Fuel_Type',
    data=df
)

plt.title("Mileage vs Selling Price")
plt.xlabel("Kilometers Driven")
plt.ylabel("Selling Price")
plt.show()
```

```python
# 6. HEATMAP - Correlation Matrix
numeric_cols = df.select_dtypes(include=['int64','float64'])

plt.figure(figsize=(8,6))
sns.heatmap(
    numeric_cols.corr(),
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
# 7. PIE CHART - Fuel Type Distribution
fuel_counts = df['Fuel_Type'].value_counts()

plt.figure(figsize=(8,8))
plt.pie(
    fuel_counts,
    labels=fuel_counts.index,
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Fuel Type Distribution")
plt.show()
```
